```python
# 다이오드의 순방향/역방향 특성 분석 예제 (상세 주석 포함 하였음)
# 이 프로그램은 다이오드 전류-전압 데이터를 분석하고, 모델을 만들고, 그래프를 그려줍니다.

# 필요한 파이썬 라이브러리들을 불러옵니다.
import numpy as np                # 수치 연산용 배열 및 수학 연산 기능
import matplotlib.pyplot as plt   # 그래프 및 시각화용
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # 한글 폰트 추가함
matplotlib.rcParams['axes.unicode_minus'] = True
#matplotlib.rcParams['font.family'] = 'Nanum Gothic'  # 유니코드 지원 폰트로 변경
#matplotlib.rcParams['axes.unicode_minus'] = True    # 유니코드 마이너스 표시 활성화
import pandas as pd               # 데이터 표 처리 및 파일 저장용

# 순방향(Forward) 다이오드 모델 함수 정의
def diode_forward_eq(V, I_s, n, V_t=0.026):
    # V: 순방향 전압 (배열 또는 숫자)
    # I_s: 포화전류 (모델의 피팅 결과)
    # n: 이상계수 (ideality factor)
    # V_t: 열전압(상온 0.026V) (default값)
    # np.minimum은 overflow 방지로 700 이하로 제한
    exponent = np.minimum(V / (n * V_t), 700)
    # 정의된 순방향 다이오드 공식: I = I_s*(exp(V/(n*V_t)) - 1)
    return I_s * (np.exp(exponent) - 1)

# 역방향(Reverse) 다이오드 모델 함수 정의
def diode_reverse_eq(V, I_s):
    # 일반적으로 역방향에서는 거의 -I_s의 값, 즉 포화전류만 흐름
    # np.ones_like(V): V와 크기가 같은 배열을 만들고 값은 -I_s로 채움
    return -I_s * np.ones_like(V)

def main():   # 프로그램 진입점 (실제 실행 내용)
    # 프로그램 시작 안내 출력
    print("=" * 65)
    print("    [다이오드 순방향/역방향 전압-전류 특성 분석 프로그램]")
    print("=" * 65)

    # 1. 순방향 데이터 준비
    # np.arange(0.0, 0.9, 0.1): 0.0 ~ 0.8V까지 0.1V 단위, 총 9개
    forward_voltages = np.arange(0.0, 0.9, 0.1)

    # 실험에서 측정한 순방향 전류(mA) 데이터. 실제 실험값으로 수정 가능!
    forward_currents = np.array([
        0.001, 0.001, 0.001, 0.001, 0.001, 0.002,
        0.02, 1.1, 8.5
    ])

    print("\n--- 순방향 데이터 (0V~0.8V) ---")
    for v, i in zip(forward_voltages, forward_currents):
        print(f"V = {v:4.1f} V | I = {i:8.3f} mA")

    # 2. 역방향 데이터 준비
    # np.arange(-28.0, 0.1, 2.0): -28V ~ 0V, 2V 간격, 총 15개
    reverse_voltages = np.arange(-28.0, 0.1, 2.0)

    # 실험에서 측정한 역방향 전류(mA) 데이터. 실제 실험값으로 수정 가능!
    reverse_currents = np.array([
        0.001, 0.001, 0.001, 0.001, 0.001, 0.001, 0.001,
        0.001, 0.001, 0.001, 0.001, 0.001, 0.001, 0.001,
        0.001
    ])

    print("\n--- 역방향 데이터 (-28V~0V) ---")
    for v, i in zip(reverse_voltages, reverse_currents):
        print(f"V = {v:5.1f} V | I = {i:8.3f} mA")

    # 3. 두 전압/전류 데이터를 하나로 합치기 (np.concatenate 사용)
    all_voltages = np.concatenate([reverse_voltages, forward_voltages])
    all_currents = np.concatenate([reverse_currents, forward_currents])

    # 4. 순방향 모델 파라미터 피팅 (로그 변환 활용)
    # 통상 다이오드는 0.6V 이상부터 지수 형태로 전류가 증가
    fit_mask = (forward_voltages >= 0.6) & (forward_currents > 0.01)
    V_fit = forward_voltages[fit_mask]
    I_fit = forward_currents[fit_mask]

    if len(V_fit) < 2:
        print('\n순방향 모델 피팅에 사용할 데이터가 부족합니다.')
        return

    # 로그 변환 후 직선 피팅: ln(I) = ln(I_s) + V/(n*V_t)
    ln_I = np.log(I_fit)
    slope, intercept = np.polyfit(V_fit, ln_I, 1)

    # 다이오드 모델 파라미터 도출
    I_s = np.exp(intercept)             # 포화전류
    n = 1 / (slope * 0.026)             # 이상계수

    # 피팅 품질(R2, 결정계수) 계산
    ln_I_pred = intercept + slope * V_fit
    ss_res = np.sum((ln_I - ln_I_pred) ** 2)        # 잔차 제곱합
    ss_tot = np.sum((ln_I - np.mean(ln_I)) ** 2)    # 총 편차 제곱합
    r_squared = 1 - (ss_res/ss_tot) if ss_tot > 0 else 0

    print("\n--- 다이오드 모델 파라미터 ---")
    print(f"I_s(포화전류): {I_s:.2e} mA")
    print(f"n(이상계수): {n:.3f}")
    print(f"결정계수 R²: {r_squared:.4f}")

    # 5. 순방향/역방향 예측값 계산
    forward_pred = diode_forward_eq(forward_voltages, I_s, n)
    reverse_pred = diode_reverse_eq(reverse_voltages, I_s)
    all_pred = np.concatenate([reverse_pred, forward_pred])

    # 예측값과 실제값의 오차 계산
    abs_err = np.abs(all_currents - all_pred)
    rel_err = 100 * abs_err / np.maximum(np.abs(all_currents), 1e-6)

    # 6. 모델 방정식 안내
    print("\n--- 도출된 방정식 ---")
    print(f"[순방향] I = {I_s:.2e} * (exp(V / ({n:.3f} * 0.026)) - 1)")
    print(f"[역방향] I ≈ -{I_s:.2e} mA (포화)")
    print("I: 전류(mA), V: 전압(V)")

    # 7. 결과 표를 DataFrame(엑셀처럼 다룸)으로 만들어서 저장
    # results = pd.DataFrame({
    #     'Voltage (V)': all_voltages,
    #     'Measured (mA)': all_currents,
    #     'Predicted (mA)': all_pred,
    #     'Abs_Error (mA)': abs_err,
    #     'Rel_Error (%)': rel_err
    # })
    # results.to_csv('diode_forward_reverse_analysis.csv', index=False, float_format='%.6f')
    # print("\n결과 CSV 파일로 저장됨: diode_forward_reverse_analysis.csv")

    # 8. 그래프 그리기 (순방향, 역방향)
    # 부드러운 예측곡선 배경용 X축 배열 만들기
    V_smooth_fwd = np.linspace(0, 0.8, 100)
    V_smooth_rev = np.linspace(-28, 0, 100)
    I_smooth_fwd = diode_forward_eq(V_smooth_fwd, I_s, n)
    I_smooth_rev = diode_reverse_eq(V_smooth_rev, I_s)


    # 두 개의 그래프 서브플롯 (순/역방향 각각)
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,6))
    ax1.scatter(forward_voltages, forward_currents, color='red', s=50, label='측정값', zorder=5)
    ax1.plot(V_smooth_fwd, I_smooth_fwd, 'b-', linewidth=2, label=f'회귀곡선 (R²={r_squared:.3f})')
    ax1.set_yscale('log')
    ax1.set_xlabel('순방향 전압 (V)')
    ax1.set_ylabel('순방향 전류 (mA)')
    ax1.set_title('다이오드 순방향 특성 (0V ~ 0.8V)')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    ax1.set_xlim(0, 0.8)
    ax1.set_ylim(0.001, max(forward_currents)*2)

    ax2.scatter(reverse_voltages, reverse_currents, color='red', s=50, label='측정값', zorder=5)
    ax2.plot(V_smooth_rev, I_smooth_rev, 'b-', linewidth=2, label='예측 곡선')
    ax2.set_xlabel('역방향 전압 (V)')
    ax2.set_ylabel('역방향 전류 (mA)')
    ax2.set_title('다이오드 역방향 특성 (-28V ~ 0V)')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    ax2.set_xlim(-28, 0)
    ax2.set_ylim(-0.001, 0.1)

#   plt.tight_layout()
#   plt.savefig('diode_forward_reverse_plot.png', dpi=300)
#   print("\n그래프 파일로 저장됨: diode_forward_reverse_plot.png")
    plt.show()

    # 9. 순+역방향 종합 그래프도 추가
    plt.figure(figsize=(12,8))
    plt.scatter(all_voltages, all_currents, color='red', s=50, label='측정값', zorder=5)
    plt.plot(V_smooth_rev, I_smooth_rev, 'b-', linewidth=2, label='역방향 예측')
    plt.plot(V_smooth_fwd, I_smooth_fwd, 'g-', linewidth=2, label='순방향 예측')
    plt.xlabel('전압 (V)')
    plt.ylabel('전류 (mA)')
    plt.title('다이오드 전체 전압-전류 특성 (-28V ~ +0.8V)')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.xlim(-28, 0.8)
    y_min = min(min(reverse_currents), min(forward_currents)) * 1.5
    y_max = max(max(reverse_currents), max(forward_currents)) * 2
    plt.ylim(y_min, y_max)
#   plt.tight_layout()
#   plt.savefig('diode_complete_characteristic.png', dpi=300)
#   print("통합 그래프 파일로 저장됨: diode_complete_characteristic.png")
    plt.show()

# 이 코드가 직접 실행될 때만 실행(main 함수 호출)
if __name__ == "__main__":
    main()
```

```python
import numpy as np

# 1. 주어진 회로 변수 설정
R = 10000        # 저항: 10k ohm
L = 33e-3        # 인덕터: 33mH
C = 0.01e-6      # 커패시터: 0.01uF
V_pp = 10        # 입력 전압: 10V 피크-투-피크

# 입력 전압의 최대치(피크값) 계산
V_max = V_pp / 2 

# 2. 공진주파수 계산 (f_r = 1 / (2 * pi * sqrt(L * C)))
f_r = 1 / (2 * np.pi * np.sqrt(L * C))

print(f"✅ 계산된 공진주파수 (f_r): {f_r:.2f} Hz\n")
print(f"{'주파수 (Hz)':<15}{'VR 최대치 (V)':<15}{'VR 실효치 (V)':<15}{'전류 I 최대치 (mA)':<18}{'전류 I 실효치 (mA)'}")
print("-" * 80)

# 3. 공진주파수 중심 및 500Hz씩 증가시켜 2000Hz까지 단계별 계산
for i in range(5):
    f = f_r + (i * 500)
    w = 2 * np.pi * f
    
    # 유도성 리액턴스 및 용량성 리액턴스 계산
    X_L = w * L
    X_C = 1 / (w * C)
    
    # 전체 임피던스 Z 계산
    Z = np.sqrt(R**2 + (X_L - X_C)**2)
    
    # 전류 I 계산 (직렬이므로 회로 전체에 동일하게 흐름)
    I_max = V_max / Z          # 전류 최대치 (A)
    I_rms = I_max / np.sqrt(2) # 전류 실효치 (A)
    
    # R에 걸리는 전압 VR 계산
    VR_max = I_max * R         # VR 최대치 (V)
    VR_rms = VR_max / np.sqrt(2) # VR 실효치 (V)
    
    # 결과 출력 (전류는 단위를 mA로 변환)
    print(f"{f:<15.2f}{VR_max:<15.4f}{VR_rms:<15.4f}{I_max*1000:<18.4f}{I_rms*1000:.4f}")
```

```python
import numpy as np

# 1. 소자 데이터 변수 지정
R = 10000        # 저항: 10k ohm
L = 33e-3        # 인덕터: 33mH
C = 0.01e-6      # 커패시터: 0.01uF
V_pp = 10        # 입력전압 피크투피크: 10V

# 입력전압 최대치(피크값) 계산
V_max = V_pp / 2 

# 2. 공진주파수 계산
f_r = 1 / (2 * np.pi * np.sqrt(L * C))

# 3. 높은 주파수가 먼저 오도록 오프셋 리스트 생성 (+2000Hz -> -2000Hz)
# 공진주파수를 중심으로 높은 쪽에서 낮은 쪽 순서대로 연산합니다.
frequency_offsets = [2000, 1500, 1000, 500, 0, -500, -1000, -1500, -2000]

print(f"✅ 계산된 공진주파수 (f_r): {f_r:.2f} Hz\n")
print(f"{'상태':<15}{'주파수 (Hz)':<15}{'VR 최대치 (V)':<15}{'VR 실효치 (V)':<15}{'전류 I 최대치 (mA)':<18}{'전류 I 실효치 (mA)'}")
print("-" * 95)

# 4. 루프를 돌며 단계별 데이터 계산 및 출력
for offset in frequency_offsets:
    # 현재 단계의 입력 주파수
    f = f_r + offset
    w = 2 * np.pi * f
    
    # 리액턴스 및 전체 임피던스 계산
    X_L = w * L
    X_C = 1 / (w * C)
    Z = np.sqrt(R**2 + (X_L - X_C)**2)
    
    # 전류 계산 (최대치 및 실효치)
    I_max = V_max / Z
    I_rms = I_max / np.sqrt(2)
    
    # R에 걸리는 전압 VR 계산 (최대치 및 실효치)
    VR_max = I_max * R
    VR_rms = VR_max / np.sqrt(2)
    
    # 상태 텍스트 서식화
    if offset > 0:
        status = f"공진 +{offset}Hz"
    elif offset < 0:
        status = f"공진 {offset}Hz"
    else:
        status = "공진 상태"
        
    # 결과 출력 (전류는 mA 단위 변환을 위해 * 1000 수행)
    print(f"{status:<15}{f:<15.2f}{VR_max:<15.4f}{VR_rms:<15.4f}{I_max*1000:<18.4f}{I_rms*1000:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# =========================================================================
# [공통 데이터 설정]
# =========================================================================
R = 10000        # 저항: 10k ohm
L = 33e-3        # 인덕터: 33mH
C = 0.01e-6      # 커패시터: 0.01uF

# 공진주파수 계산
f_r = 1 / (2 * np.pi * np.sqrt(L * C))
w_r = 2 * np.pi * f_r
frequency_offsets = [2000, 1500, 1000, 500, 0, -500, -1000, -1500, -2000]

print(f"✅ 계산된 공진주파수 (f_r): {f_r:.2f} Hz")


# =========================================================================
# [PART 1] 직렬 공진회로 분석 (입력 전압: 10V p-p)
# =========================================================================
V_pp_1 = 10
V_max_1 = V_pp_1 / 2 

calc_freqs_1 = []
calc_currents_max_1 = []
calc_phases_1 = []

print("\n" + "="*45 + " [실험 1] 직렬 공진회로 결과 " + "="*45)
print(f"{'상태':<13}{'주파수 (Hz)':<14}{'VR 최대치 (V)':<14}{'전류 I 최대 (mA)':<16}{'위상각 (Degree)':<15}")
print("-" * 105)

for offset in frequency_offsets:
    f = f_r + offset
    w = 2 * np.pi * f
    X_L = w * L
    X_C = 1 / (w * C)
    
    # 직렬 임피던스 (복소수 형태)
    Z_series_complex = R + 1j * (X_L - X_C)
    Z_mag = np.abs(Z_series_complex)
    
    # 위상각 계산 (전류 I의 위상 = - 임피던스의 위상)
    phase_rad = -np.angle(Z_series_complex)
    phase_deg = np.degrees(phase_rad)
    
    I_max = V_max_1 / Z_mag
    VR_max = I_max * R
    
    calc_freqs_1.append(f)
    calc_currents_max_1.append(I_max * 1000)
    calc_phases_1.append(phase_deg)
    
    status = f"공진 +{offset}Hz" if offset > 0 else (f"공진 {offset}Hz" if offset < 0 else "공진 상태")
    print(f"{status:<13}{f:<14.2f}{VR_max:<14.4f}{I_max*1000:<16.4f}{phase_deg:<15.4f}")


# =========================================================================
# [PART 2] 병렬 공진회로 분석 (입력 전압: 4V p-p, R과 (L//C)의 직렬)
# =========================================================================
V_pp_2 = 4
V_max_2 = V_pp_2 / 2

calc_freqs_2 = []
calc_currents_max_2 = []
calc_phases_2 = []

print("\n" + "="*45 + " [실험 2] 병렬 공진회로 결과 " + "="*45)
print(f"{'상태':<13}{'주파수 (Hz)':<14}{'VR 최대치 (V)':<14}{'전류 I 최대 (mA)':<16}{'위상각 (Degree)':<15}")
print("-" * 105)

for offset in frequency_offsets:
    f = f_r + offset
    w = 2 * np.pi * f
    
    # 병렬 LC 어드미턴스 및 임피던스 계산
    Y_LC = 1 / (1j * w * L) + 1j * w * C
    Z_LC = 1 / Y_LC if not np.isclose(Y_LC, 0) else np.inf
    
    # 전체 임피던스 (복소수)
    Z_total_complex = R + Z_LC
    
    if np.isinf(Z_total_complex.real) or np.isinf(Z_total_complex.imag):
        I_max = 0.0
        phase_deg = 0.0  # 공진점에서는 전류가 0이 되므로 위상차가 정의되지 않으나 극한값은 0으로 수렴
    else:
        Z_mag = np.abs(Z_total_complex)
        I_max = V_max_2 / Z_mag
        phase_rad = -np.angle(Z_total_complex)
        phase_deg = np.degrees(phase_rad)
        
    VR_max = I_max * R
    
    calc_freqs_2.append(f)
    calc_currents_max_2.append(I_max * 1000)
    calc_phases_2.append(phase_deg)
    
    status = f"공진 +{offset}Hz" if offset > 0 else (f"공진 {offset}Hz" if offset < 0 else "공진 상태")
    print(f"{status:<13}{f:<14.2f}{VR_max:<14.4f}{I_max*1000:<16.4f}{phase_deg:<15.4f}")


# =========================================================================
# [PART 3] Q값 및 대역폭 계산 출력
# =========================================================================
print("\n" + "="*38 + " [PART 3] Q-Factor 및 대역폭 분석 결과 " + "="*38)
Q_series = (w_r * L) / R
BW_series = f_r / Q_series
print(" 1) 직렬 공진회로 : " + f"Q = {Q_series:.4f}  |  BW = {BW_series:.2f} Hz")

Q_parallel = R / (w_r * L)
BW_parallel = f_r / Q_parallel
print(" 2) 병렬 공진회로 : " + f"Q = {Q_parallel:.4f}  |  BW = {BW_parallel:.2f} Hz")
print("=" * 105)


# =========================================================================
# [PART 4] 전류 크기 및 위상각 연속 곡선 데이터 생성 및 시각화
# =========================================================================
f_continuous = np.linspace(min(calc_freqs_1) - 1000, max(calc_freqs_1) + 1000, 2000)
w_cont = 2 * np.pi * f_continuous

# 1) 직렬 연속 복소수 데이터
Z_cont_s_comp = R + 1j * (w_cont * L - 1 / (w_cont * C))
I_cont_s = (V_max_1 / np.abs(Z_cont_s_comp)) * 1000
phase_cont_s = np.degrees(-np.angle(Z_cont_s_comp))

# 2) 병렬 연속 복소수 데이터
Y_cont_p_comp = 1 / (1j * w_cont * L) + 1j * w_cont * C
Z_cont_p_comp = R + 1 / Y_cont_p_comp
I_cont_p = (V_max_2 / np.abs(Z_cont_p_comp)) * 1000
phase_cont_p = np.degrees(-np.angle(Z_cont_p_comp))

# 그래프 출력 기동 (2행 2열 구조)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

# --- [상단] 전류 크기 그래프 ---
# 직렬 전류 크기
ax1.plot(f_continuous, I_cont_s, color='#2ca02c', linestyle='--', label='Continuous Curve')
ax1.scatter(calc_freqs_1, calc_currents_max_1, color='red', s=40, zorder=5)
ax1.axvline(x=f_r, color='blue', linestyle=':', alpha=0.7, label=f'f_r = {f_r:.1f}Hz')
ax1.set_title(f'1-A) Series Circuit: Current Magnitude', fontsize=11, fontweight='bold')
ax1.set_ylabel('Current, I_max (mA)')
ax1.grid(True, linestyle=':', alpha=0.6)
ax1.legend(loc='upper right')

# 병렬 전류 크기
ax2.plot(f_continuous, I_cont_p, color='#1f77b4', linestyle='--', label='Continuous Curve')
ax2.scatter(calc_freqs_2, calc_currents_max_2, color='red', s=40, zorder=5)
ax2.axvline(x=f_r, color='blue', linestyle=':', alpha=0.7, label=f'f_r = {f_r:.1f}Hz')
ax2.set_title(f'2-A) Parallel Circuit: Current Magnitude', fontsize=11, fontweight='bold')
ax2.set_ylabel('Current, I_max (mA)')
ax2.grid(True, linestyle=':', alpha=0.6)
ax2.legend(loc='lower right')

# --- [하단] 위상각 그래프 (Phase Angle) ---
# 직렬 위상각
ax3.plot(f_continuous, phase_cont_s, color='orange', linestyle='-', label='Phase Curve')
ax3.scatter(calc_freqs_1, calc_phases_1, color='red', s=40, zorder=5)
ax3.axvline(x=f_r, color='blue', linestyle=':', alpha=0.7)
ax3.axhline(y=0, color='black', linestyle='-', alpha=0.3)
ax3.set_title('1-B) Series Circuit: Phase Angle (Current vs V_in)', fontsize=11, fontweight='bold')
ax3.set_xlabel('Frequency, f (Hz)')
ax3.set_ylabel('Phase Angle (Degrees)')
ax3.grid(True, linestyle=':', alpha=0.6)
ax3.legend(loc='upper right')

# 병렬 위상각
ax4.plot(f_continuous, phase_cont_p, color='purple', linestyle='-', label='Phase Curve')
ax4.scatter(calc_freqs_2, calc_phases_2, color='red', s=40, zorder=5)
ax4.axvline(x=f_r, color='blue', linestyle=':', alpha=0.7)
ax4.axhline(y=0, color='black', linestyle='-', alpha=0.3)
ax4.set_title('2-B) Parallel Circuit: Phase Angle (Current vs V_in)', fontsize=11, fontweight='bold')
ax4.set_xlabel('Frequency, f (Hz)')
ax4.set_ylabel('Phase Angle (Degrees)')
ax4.grid(True, linestyle=':', alpha=0.6)
ax4.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
