```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# 한글 폰트 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False

# -------------------------------
# 입력 신호 설정
# -------------------------------
t = np.linspace(0, 0.01, 1000)   # 시간 (10ms 구간)
f = 1000                         # 주파수 1kHz
V1 = 4 * np.sin(2 * np.pi * f * t)   # 입력 전압 1 (4V)
V2 = 3 * np.sin(2 * np.pi * f * t)   # 입력 전압 2 (3V)

# -------------------------------
# 회로 파라미터
# -------------------------------
R1 = 10000     # V1 입력 저항 [Ω]
R2 = 10000     # V2 입력 저항 [Ω]
Rf = 10000     # 피드백 저항 [Ω]

# -------------------------------
# 1️⃣ 가산 증폭기 (Summing Amplifier)
# -------------------------------
# 출력: Vout = -Rf * (V1/R1 + V2/R2)
Vout_sum = -Rf * ((V1 / R1) + (V2 / R2))

# -------------------------------
# 2️⃣ 감산 증폭기 (Differential Amplifier)
# -------------------------------
# 출력: Vout = (Rf/R1)*(V2 - V1)
Vout_sub = (Rf / R1) * (V2 - V1)

# -------------------------------
# 그래프 표시
# -------------------------------
plt.figure(figsize=(12,8))

# 입력 전압 1, 2
plt.subplot(3,1,1)
plt.plot(t, V1, label="입력 전압 V1 (4V)")
plt.plot(t, V2, label="입력 전압 V2 (3V)")
plt.title("입력 전압 V1, V2 (정현파)")
plt.ylabel("전압 [V]")
plt.grid(True)
plt.legend()

# 가산 회로 출력
plt.subplot(3,1,2)
plt.plot(t, Vout_sum, 'r', label="가산 회로 출력 (Summing Amplifier)")
plt.title("가산 회로 출력")
plt.ylabel("전압 [V]")
plt.grid(True)
plt.legend()

# 감산 회로 출력
plt.subplot(3,1,3)
plt.plot(t, Vout_sub, 'g', label="감산 회로 출력 (Differential Amplifier)")
plt.title("감산 회로 출력")
plt.xlabel("시간 [s]")
plt.ylabel("전압 [V]")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
