```python
import numpy as np
import matplotlib.pyplot as plt

# 기본 파라미터
Vcc = 12          # 전원전압 (V)
Rc = 2e3          # 컬렉터 저항 (2kΩ)
beta = 100        # 트랜지스터 전류 이득 (β)
Rin = 10e3        # 베이스 입력 저항 (10kΩ)

# 시간축
fs = 100e3
t = np.arange(0, 2e-3, 1/fs)  # 2ms 구간

# 입력 신호 (AC + DC)
Vin_ac = 0.02 * np.sin(2 * np.pi * 1e3 * t)  # 20mV 사인파
Vb_dc = 0.7                                  # 베이스 바이어스 (V_BE 기준)
Vb = Vb_dc + Vin_ac

# 베이스 전류 (단순 저항 입력으로 가정)
Ib = Vin_ac / Rin

# 컬렉터 전류
Ic = beta * Ib

# 출력 전압
Vout = Vcc - (Ic * Rc)

# 시각화
plt.figure(figsize=(10, 8))

# 입력 베이스 신호
plt.subplot(3, 1, 1)
plt.plot(t * 1000, Vin_ac * 1000, color='blue')
plt.title('Input Signal (AC component at Base)')
plt.ylabel('Voltage (mV)')
plt.grid(True)

# 컬렉터 전류
plt.subplot(3, 1, 2)
plt.plot(t * 1000, Ic * 1e3, color='red')
plt.title('Collector Current (Ic)')
plt.ylabel('Current (mA)')
plt.grid(True)

# 출력 전압
plt.subplot(3, 1, 3)
plt.plot(t * 1000, Vout, color='green')
plt.title('Output Voltage at Collector (Vout)')
plt.xlabel('Time (ms)')
plt.ylabel('Voltage (V)')
plt.grid(True)

plt.tight_layout()
plt.show()

```
